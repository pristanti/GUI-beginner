```
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())

worksheet = gc.open('Go Play - Assignment').sheet1
```

```
# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
```

```
# Convert to a DataFrame and render.
import pandas as pd
tableA = pd.DataFrame.from_records(rows, columns=rows[0])
tableA.drop(index=0, axis=1, inplace=True)
tableA.head(10)
```

```
tableA["Total CBV"] = tableA["Total CBV"].str.replace("-", "0")
tableA.head(12)
```

```
# create, and save df
from gspread_dataframe import set_with_dataframe
title = 'Go Play - Assignment-2'
gc.create(title)
sheet = gc.open(title).sheet1
set_with_dataframe(sheet, tableA) 
tableA
```
